```
from abc import ABC, abstractmethod

class Chair(ABC):
  @abstractmethod
  def release_chair(self):
    pass

class VictorianChair(Chair):
  def release_chair(self):
    print('создаем викторианский стул')

class ModernChair(Chair):
  def release_chair(self):
    print('создаем модернистский стул')


class CoffeTable(ABC):
  @abstractmethod
  def release_CoffeTable(self):
    pass

class VictorianCoffeTable(CoffeTable):
  def release_CoffeTable(self):
    print('создаем викторианский кофейный столик')

class ModernCoffeTable(CoffeTable):
  def release_CoffeTable(self):
    print('создаем модернистский кофейный столик')



class Sofa(ABC):
  @abstractmethod
  def release_Sofa(self):
    pass

class VictorianSofa(Sofa):
  def release_Sofa(self):
    print('создаем викторианский диван')

class ModernSofa(Sofa):
  def release_Sofa(self):
    print('создаем модернистский диван')



class IFactory(ABC):
  @abstractmethod
  def create_chair(self) -> Chair:
    pass

  @abstractmethod
  def create_coffe_table(self) -> CoffeTable:
    pass

  @abstractmethod
  def create_sofa(self) -> Sofa:
    pass


class VictorianFurnitureFactory(IFactory):
  def create_chair(self) -> Chair:
    return VictorianChair()

  def create_coffe_table(self) -> CoffeTable:
    return VictorianCoffeTable() 

  def create_sofa(self) -> Sofa:
    return VictorianSofa()

class ModernFurnitureFactory(IFactory):
  def create_chair(self) -> Chair:
    return ModernChair()

  def create_coffe_table(self) -> CoffeTable:
    return ModernCoffeTable() 

  def create_sofa(self) -> Sofa:
    return ModernSofa() 

V_factory = VictorianFurnitureFactory()
V_sofa = V_factory.create_sofa()
V_sofa.release_Sofa()

M_factory = ModernFurnitureFactory()
M_sofa = M_factory.create_chair()
M_sofa.release_chair()
```

```
class Car:
  def __init__(self):
    self.data: str = ''

  def about_car(self) -> str:
    return self.data

  def append_data(self, string: str):
    self.data += string

class IDeveloper(ABC):
  def create_body(self):
    pass

  def crete_wheels(self):
    pass

  def create_engine(self):
    pass

  def install_navigation_system(self):
    pass

  def create_air_conditioner(self):
    pass

  def get_car(self) ->Car:
    pass

class BMW(IDeveloper):
  def __init__(self):
    self.__car = Car()

  def create_body(self):
    self.__car.append_data("произведен корпус BMW; ")

  def crete_wheels(self):
    self.__car.append_data("произведены колеса BMW; ")
  
  def create_engine(self):
    self.__car.append_data("проведен двиатель BMW; ")

  def install_navigation_system(self):
    self.__car.append_data("установлена навиационная система; ")

  def create_air_conditioner(self):
    self.__car.append_data("установлен кондиционер; ")

  def get_car(self) -> Car:
    return self.__car

    
class HONDA(IDeveloper):
  def __init__(self):
    self.__car = Car()

  def create_body(self):
    self.__car.append_data("произведен корпус HONDA; ")

  def crete_wheels(self):
    self.__car.append_data("произведены колеса HONDA; ")
  
  def create_engine(self):
    self.__car.append_data("проведен двиатель HONDA; ")

  def install_navigation_system(self):
    self.__car.append_data("установлена навиационная система; ")

  def create_air_conditioner(self):
    self.__car.append_data("установлен кондиционер; ")

  def get_car(self) -> Car:
    return self.__car


class Director:
  def __init__(self,developer: IDeveloper):
    self.__developer =  developer

  def set_developer(self, developer: IDeveloper):
    self.__developer = developer

  def create_without_navigator(self)-> Car:
    self.__developer.create_body()
    self.__developer.crete_wheels()
    self.__developer.create_engine()
    self.__developer.create_air_conditioner()
    return self.__developer.get_car()

  def create_full_complectation(self) -> Car:
    self.__developer.create_body()
    self.__developer.crete_wheels()
    self.__developer.create_engine()
    self.__developer.install_navigation_system()
    self.__developer.create_air_conditioner()
    return self.__developer.get_car()


honda_factory = HONDA()
director = Director(honda_factory)
full_comp_honda= director.create_full_complectation()
print(full_comp_honda.about_car())
bmw_factory = BMW()
director = Director(bmw_factory)
bmw_car = director.create_without_navigator()
print(bmw_car.about_car())
   
```
